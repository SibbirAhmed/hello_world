```python
for index,row in enumerate(data_col):
    if row == 'yearOfRegistration':
        data_col[index]= 'registration_year'
        
    elif row == 'monthOfRegistration':
        data_col[index]= 'registration_month'
        
    elif row == 'notRepairedDamage':
        data_col[index]= 'unrepaired_damage'
        
    elif row == 'dateCreated':
        data_col[index]= 'ad_created'
    
    else:
        count=-1
        for char in row:
            count += 1
            if ord(char) in range(65,90):
                row=row[:count] + '_' + row[count:]
                count += 1
       
        row=row.lower()
        data_col[index]= row
        
print(data_col)
               
        
```

```python
booli= autos['brand'].value_counts(normalize=True) > 0.05
```

```python
print(booli)
```

we selected brands that have over a
certain percentage of the total values (e.g. > 5%). and those are 
volkswagen        0.212898
bmw               0.108507

```python
brand_mean_price={}
count=0
for name in booli:
    if name:
        brand_name=booli.index[count]
        all_row_of_that_brand_name= autos['brand'] == brand_name
        brand_mean_price[brand_name]=autos.loc[all_row_of_that_brand_name,'price'].mean()
    count += 1
print(brand_mean_price)
```

```python
brand_mean_mileage={}
count=0
for name in booli:
    if name:
        brand_name=booli.index[count]
        all_row_of_that_brand_name= autos['brand'] == brand_name
        brand_mean_mileage[brand_name]=autos.loc[all_row_of_that_brand_name,'odometer_km'].mean()
    count += 1
print(brand_mean_mileage)
```

```python
bmp_series= pd.Series(brand_mean_price)
print(bmp_series)
```

```python
bmm_series= pd.Series(brand_mean_mileage)
print(bmm_series)
```

```python
df= pd.DataFrame(bmp_series,columns=['mean_price'])
df
```

```python
df.loc[:,'mean_mileage']= bmm_series
df
```
