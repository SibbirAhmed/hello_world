### Identify dissatisfied Employees

```python
#counting dissatisfied person number in a dissatisfaction cell of tafe resignations
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
#counting dissatisfied person number in job dissatisfaction cell of tafe resignations
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
# function to decide if the job was left because of disstisfaction(true) or not or it is null
def update_vals(value):
    if pd.isnull(value):
        return np.nan
    elif value == '-':
        return False
    else:
        return True
```

```python
# create dissatisfied column in tafe_registration using applymap and any methods and copy the result
tafe_resignations['dissatisfied']=tafe_resignations[['Contributing Factors. Dissatisfaction','Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(axis=1,skipna=False)
tafe_resignations_up=tafe_resignations.copy()
tafe_resignations['dissatisfied'].value_counts(dropna=False)
```

```python
diss_col=['job_dissatisfaction',
'dissatisfaction_with_the_department',
'physical_work_environment',
'lack_of_recognition',
'lack_of_job_security',
'work_location',
'employment_conditions',
'work_life_balance',
'workload']
```

```python
# create dissatisfied column in dete_registration using any() and copy the result
dete_resignations['dissatisfied']=dete_resignations[diss_col].any(axis=1,skipna=False)
dete_resignations_up=dete_resignations.copy()
dete_resignations['dissatisfied'].value_counts(dropna=False)
```

#### 91 people left job in tafe because of dissatisfaction and 149 people in dete

### combine tha data

```python
# create 'institute' column for both dataframe
dete_resignations_up['institute']="DETE"
tafe_resignations_up['institute']="TAFE"
```

```python
# combine dataframe
combined=pd.concat([dete_resignations_up,tafe_resignations_up],ignore_index=True)
combined.head()
```

```python
# checking number of notnull value
combined.notnull().sum().sort_values()
```

```python
# drop columns has less than 500 notnull values
combined_updated= combined.dropna(thresh=500 , axis=1)
combined_updated.info()
```

### clean the service column

```python
# check institute_service
combined_updated['institute_service'].value_counts(dropna=False)
```

```python
# extract all years and convert into float
combined_updated['institute_service_up']=combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
combined_updated['institute_service_up'].value_counts()
```

```python
# function to category job experience
def category(val):
    if pd.isnull(val):
        return np.nan
    elif 1 <= val <=2 :
        return "New"
    elif 3 <= val <= 6:
        return "Experienced"
    elif 7 <= val <= 10:
        return "Established"
    else:
        return "Veteran"
```

```python
# create "service_cat" column applying above function
combined_updated['service_cat']= combined_updated['institute_service_up'].apply(category)
combined_updated['service_cat'].value_counts()
```

### perform initial analysis

```python
# checking dissatisfied value
combined_updated["dissatisfied"].value_counts(dropna= False)
```

```python
# fill null values with False
combined_updated['dissatisfied']=combined_updated['dissatisfied'].fillna(False)
combined_updated["dissatisfied"].value_counts(dropna= False)
```

```python
# category-wise dissatisfaction percentage
diss_prcntge=combined_updated.pivot_table('dissatisfied','service_cat')
diss_prcntge
```

```python
#plot the dissatisfaction percentage
%matplotlib inline
diss_prcntge.plot(kind='bar')
```

## From  the plot we can say that , the established category who had job experience between 7 to 10 years, left job most due to some kind of dissatisfaction.
